## The point of this code is to found out which Citi Bike docks are used the most by Customers. These are temporary users such as tourists. From my previous work, this set of users has room to grow so they could be targeted by knowing which bikes they use.

### As you can see by the graph at the bottom (or in the graph file) that the most used docks from 2013 by tourists are the ones by central park and at  brooklyn bridge



```python
#libraries needed
import csv
import pandas as pd
import os

## this is the datetime class that has useful functions to manipulate dates.
from datetime import datetime

#these are the plotly libaries, but i dont use ff
from plotly.tools import FigureFactory as ff
import plotly
import plotly.graph_objs as go

plotly.offline.init_notebook_mode(connected=True)
```

### First thing I did was to import each file, read the data and put it in a df, then I indexed it by user type so i could separate out the Customers.


```python
#I needed to rename the columns so they is consistent amoung files. This is just the list of column titles to use later
columnlist=['Trip Duration', 'Start Time','Stop Time', 'Start Station ID', 'Start Station Name', 
            'Start Latitude', 'Start Longitude', ' End Station ID', 'End Station Name', 'End Latitude', 
            'End Longitude', 'Bike ID', 'User Type', 'Birth Year', 'Gender']

# import data directory the . at the start is the location of the python file
import_directory='./trip_data_monthly/'

#blank dataframe to store the data in
df_all_cust=pd.DataFrame()

#iterate over files in a directory
for file in os.listdir(import_directory):
    if file.endswith(".csv"):# this was important as their were hidden files in the folder that messed with my code further down
        #reads the CSV into a dataframe
        df_month = pd.read_csv(import_directory + file)
        
        print(file)# for my own info of which files are being read
       
        # rename the columns
        df_month.columns= columnlist
        
        #index the dataframe by user type
        df_month_ind=df_month.set_index("User Type")
        
        #subset the dataframe by Customers from the index
        df_month_cust=df_month_ind.loc['Customer']
        

        #append the dataframe to the all users data frame (this is why having consistent column names is important)
        df_all_cust=df_all_cust.append(df_month_cust)
        
# printed just to see the shape of the dataframe
print(df_all_cust.shape)
        
```

```python
#looking at the head of the dataframe just to confirm Customer sort worked ok
df_all_cust.head()
```

### Now I can group the data frame by longitude and latitude of the docks. This will give me the count/size/ number of times a bike has been removed from a dock at a particular position

```python
# grouping by two columns here, i shouldn't actually need to use two columns because each unique latitude should 
#have a unique longitude. the results are save to a new dataframe
df_positions= df_all_cust.groupby(['Start Latitude', 'Start Longitude']).size().reset_index()

# rename colums for new dataframe
df_positions.columns=['Start Latitude', 'Start Longitude', 'Count']
df_positions.head()
```

```python
# sorted dataframe by count just to see the values
df_position_sort=df_positions.sort_values(by=['Count'])
df_position_sort.head()
```

### now I am ready to plot the data using a scatter graph with colour code depending on use

```python
#these are the coordinates of central park and brooklyn bridge so i can mao them onto my graph
central_park_y=[40.76805, 40.76428, 40.79687, 40.80057, 40.76805]
central_park_x= [-73.98189, -73.97301, -73.94923, -73.95818,-73.98189]

brookbrid_y=[40.71244, 40.70053]
brookbrid_x=[-74.00489, -73.98986]

```

```python
# plotting the graph using plotly

#1 define data in [] brackets
data_loc=[go.Scatter(x= df_positions['Start Longitude'], y=df_positions['Start Latitude'],name='Docks',  mode='markers',
        marker=dict(
        size=10,
        color = df_positions['Count'], #set color equal to a variable
        colorscale='Jet',
        showscale=True, 
        opacity=0.9)), 
        go.Scatter(x= central_park_x, y = central_park_y, mode='lines+markers', name = 'Central Park',
        opacity=0.5,  marker=dict(size=3, color = 'rgb(255, 255, 255)', opacity=0.5), line = dict(width = 3)),
        go.Scatter(x= brookbrid_x, y = brookbrid_y, mode='lines+markers', name = 'Brooklyn Bridge',
        opacity=0.5,  marker=dict(size=3, color = 'rgb(255, 255, 255)', opacity=0.), line = dict(width = 3))
         ]




#2 adjust layout of graph
layout_locations = go.Layout(
    xaxis=dict(range=[-74.02, -73.92], title="Longitude"),
    yaxis=dict(range=[40.65, 40.83], title= "Latitude"),
    plot_bgcolor='black',
    title=go.layout.Title(text='Docks used by Customer', font=dict(size=18)))

# 3 define figure
fig_locations = go.Figure(data=data_loc, layout=layout_locations)

# 4 plot in notebook
plotly.offline.iplot(fig_locations, filename='Rides.html')
```
